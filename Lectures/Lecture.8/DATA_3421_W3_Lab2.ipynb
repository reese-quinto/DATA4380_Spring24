### **5. Format Inconsistencies**

**Cleaning and Handling Format Inconsistencies**

1. **Categorical Data:**
Standardize the text format by converting all strings to the same case (usually lower or upper) and trimming unnecessary whitespace.

2. **Datetime Data:**
Convert all date/time entries to a standard format. This might require using pd.to_datetime with a specific format string.

```python
# Function to clean format inconsistencies
def clean_format_inconsistencies(column, data_type):
    if data_type == 'categorical':
        # Convert to lower case and strip whitespace
        df[column] = df[column].str.lower().str.strip()
    elif data_type == 'datetime':
        # Convert to a standard datetime format
        df[column] = pd.to_datetime(df[column], errors='coerce', format='%Y-%m-%d')  # Specify the format as per your data

# Apply the cleaning function for each column
for column in df.columns:
    data_type = 'categorical' if pd.api.types.is_categorical_dtype(df[column]) else 'datetime' if pd.api.types.is_datetime64_any_dtype(df[column]) else 'other'
    if data_type in ['categorical', 'datetime']:
        clean_format_inconsistencies(column, data_type)

# Display the first few rows of the cleaned dataframe
print(df.head())
```


**Here, for categorical columns:**
- I converts all strings to lowercase and strips any leading/trailing whitespace.

**For datetime columns:**
 - it standardizes the date format. You should adjust the format parameter in pd.to_datetime to match the format of your date strings.

This process will help ensure consistency in your dataset, making it more reliable for analysis.

### **8. Check for Outliers**

**Ready!!!**

The next step involves examining numeric columns for outliers or other anomalies that might need attention.

This includes columns such as Income, Height, Credit Score, and Debt. We will identify outliers using a simple statistical method based on the interquartile range (IQR).

This method is commonly used for outlier detection in datasets. **Lets start**

```python

# Step 8: Examine numeric columns for outliers

# Lets define a function to detect outliers based on the IQR method
def detect_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (column < lower_bound) | (column > upper_bound)

# Apply outlier detection
outliers_income = detect_outliers(df['Income']).sum()
outliers_height = detect_outliers(df['Height']).sum()
outliers_credit_score = detect_outliers(df['Credit Score']).sum()
outliers_debt = detect_outliers(df['Debt']).sum()

outliers_summary = {
    'Income Outliers': outliers_income,
    'Height Outliers': outliers_height,
    'Credit Score Outliers': outliers_credit_score,
    'Debt Outliers': outliers_debt
}

outliers_summary

```

**Handling outliers depends on the context of your analysis and the specific reasons these values are outliers. Options include:**

- **Leaving them as is:** If the outliers represent valid data points that are important for the analysis.


- **Removing them:** If they are considered to be errors or not representative of the data you want to analyze.

- **Capping them:** Replacing outlier values with the nearest value within the acceptable range (either the lower or upper bound calculated during outlier detection).

## **Handling Outliers**


**To address the outliers in the Debt and Income columns using different techniques, I will use these two methods. But depend on your dataset you can use other methods too**

- For **Income outliers**, we'll cap them, replacing outlier values with the nearest value within the acceptable range. This approach maintains the presence of these entries in the dataset while reducing the impact of extreme values.

- For **Debt outliers**, considering its lower number, we might opt for a more conservative approach like removing these rows entirely, assuming they could represent data entry errors or extreme cases not representative of the general population.

```python
# Define bounds for capping Income outliers
Q1_income = df['Income'].quantile(0.25)
Q3_income = df['Income'].quantile(0.75)
IQR_income = Q3_income - Q1_income
lower_bound_income = Q1_income - 1.5 * IQR_income
upper_bound_income = Q3_income + 1.5 * IQR_income

# Cap 'Income' outliers
df['Income'] = df['Income'].clip(lower=lower_bound_income, upper=upper_bound_income)
```

```python
# For 'Debt', remove outliers
# Calculate IQR for Debt
Q1_debt = df['Debt'].quantile(0.25)
Q3_debt = df['Debt'].quantile(0.75)
IQR_debt = Q3_debt - Q1_debt
lower_bound_debt = Q1_debt - 1.5 * IQR_debt
upper_bound_debt = Q3_debt + 1.5 * IQR_debt

# Identify indices of Debt outliers
outlier_indices_debt = df[(df['Debt'] < lower_bound_debt) | (df['Debt'] > upper_bound_debt)].index

# Remove 'Debt' outliers
data_cleaned = df.drop(index=outlier_indices_debt)

```

```python

# Display changes: Summarize the current state of 'Income' and 'Debt', and confirm 'Debt' correction
summary_changes = {
    'Rows After Debt Outlier Removal': data_cleaned.shape[0],
    'Corrected Column Name': 'Debt' in data_cleaned.columns
}

summary_changes, data_cleaned[['Income', 'Debt']].describe()
```

### **Normalization**

Normalization and standardization are crucial preprocessing steps, particularly for machine learning models sensitive to the scale of the data. There are several methods to scale your data:

- **Normalization (Min-Max Scaling):** Transforms features to be on a scale from 0 to 1.
- **Mean Normalization:**Scales data to have a mean of 0 and a range between -1 and 1.
- **Standardization (Z-score Normalization):** Transforms data to have a mean of 0 and a standard deviation of 1.
Let's apply these methods to your dataset:

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler


# Separate numerical and categorical columns
numerical_columns = df.select_dtypes(include=['number']).columns

# Apply Min-Max Scaling to numerical columns between 0 and 1
scaler_minmax = MinMaxScaler()
df[numerical_columns] = scaler_minmax.fit_transform(df[numerical_columns])
df.head()
```

```python

# Apply Standardization to numerical columns
scaler_standard = StandardScaler()
df[numerical_columns] = scaler_standard.fit_transform(df[numerical_columns])

print("Data with scaled numerical columns between 0 and 1:")
print(df.head())
```

Before applying these transformations, ensure that you've handled missing values and outliers, as they can significantly affect the scaling.
Remember to apply the same scaling to new data (e.g., test data in machine learning) using the parameters learned from the training data.
The choice between these methods depends on the requirements of your analysis or model. For instance, some models require standardization over normalization and vice versa.
It's common to try different scaling methods and compare model performance to decide which is the best for your specific case.
